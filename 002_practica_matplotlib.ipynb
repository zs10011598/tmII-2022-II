### Grafica de una funcion $f: R → R$

```python
fig, ax = plt.subplots() # define Figure y Axes

x = np.linspace(0, 10, 100) # definir los puntos del dominio
y = np.cos(x) # aplicar la funcion a los puntos del dominio

ax.plot(x, y)     # graficar la funcion 
#ax.scatter(x, y) # graficas los puntos
#ax.fill(x, y)    # graficacion rara
```

### Grafica parametrizada $f: R → R^3$

```python
fig = plt.figure()              # funcion para definir una grafica sin ejes
ax = plt.axes(projection='3d')  # definir ejes de R3 en la figura 

t = np.linspace(0, 15, 1000)

z = t
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, z, 'gray') # graficar en 3D
```

### Campos vectoriales $F: R^2 → R^2 $, $F(x, y) = (u, v)$

```python
x, y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10)) # particion en el plano

u = -3
v = -1

plt.quiver(x, y, u, v) # grafica un campo vectorial
```

### Campos vectoriales $F: R^3 → R^3 $, $F(x, y, z) = (u, v, w)$

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x,  y, z = np.meshgrid(np.linspace(-0.8, 1, 10), np.linspace(-0.8, 1, 10), np.linspace(-0.8, 1, 3))

u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) * np.sin(np.pi * z)

ax.quiver(x, y, z, u, v, w, length=0.1, color='black')
```

### Curvas de nivel $F: R^2 → R$

```python
def F(x, y):
  return np.sin(np.sqrt(x**2 + y **2))

x, y = np.meshgrid(np.linspace(-6, 6, 30), np.linspace(-6, 6, 30))
z = F(x, y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour(x, y, z, 20)
```
